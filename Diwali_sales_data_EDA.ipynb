<h1>EXPLANATORY DATA ANALYSIS</h1>

```python
#  PLOTTING A BAR CHART FOR GENDER AND ITS COUNT
# HERE ax IS A VARIABLE ASSIGNED WITH THE SEABORN COUNTPLOT OF GENDER
# ax.containers IS A LIST WHICH HOLDS ALL THE BAR GROUPS. HERE ONLY HOLDS GENDER GROUP
#ax.bar_label() THIS IS A ACTION COMMAND WHICH TAKES ALL THE GROUP OF BARS AND CALCULATE ITS HEIGHT
# (bars) THIS ACTUALLY CURRENT GROUP OF BARS THAT LOOP GRABS FROM FIRST LINE i.e.  ax.containers
# FOR LOOP IS USED

sns.set(rc={"figure.figsize":(8,5)})
ax=sns.countplot(x="Gender", data=df)

for bars in ax.containers :
    ax.bar_label(bars)
```

```python
# PLOTTING A BAR CHART FOR GENDER VS TOTAL AMOUNT


#Part-by-Part Breakdown

#  df: This is your starting point. It represents your main data table (DataFrame) containing all your raw sales information.
# .groupby(['Gender'], as_index=False): This splits your data into separate piles based on gender (e.g., a Male pile and a Female pile).
# The as_index=False prevents "Gender" from becoming a bold row label, keeping it as a normal, easy-to-use column instead.
# ['Amount']: This tells the code to look only at the money column inside those gender piles. It ignores irrelevant columns like names, dates, or ID numbers.
# .sum(): This adds together every single number in the "Amount" column for each gender pile, creating a total sum for each group.
# .sort_values(by='Amount', ascending=False): This ranks your results. It looks at the new total amounts and sorts them from highest to lowest (ascending=False).


sales_gen = df.groupby(["Gender"], as_index=False)["Amount"].sum().sort_values(by="Amount", ascending=False)

ax=sns.barplot(x="Gender", y="Amount", data=sales_gen)

for bars in ax.containers :
    ax.bar_label(bars)
```

<b> The above graph shows that most number of buyers are women as well as they have most purchasing power.</b>

```python
# NUMBER OF MALE AND FEMALE IN EACH AGE GROUP
ax = sns.countplot(x="Age Group", hue="Gender", data= df)

for bars in ax.containers :
    ax.bar_label(bars)
```

```python
# TOTAL AMOUNT VS AGE GROUP
#  THIS GRAPH SHOWS THAT AMOUNT OF MONEY SPENT BY EACH AGE GROUP

df_new= df.groupby(["Age Group"], as_index=False)["Amount"].sum().sort_values(by="Amount", ascending=False)

ax = sns.barplot(x="Age Group", y="Amount", data=df_new)

for bars in ax.containers :
    ax.bar_label(bars)
```

```python
# TOTAL ORDERS VS AGE GROUP 
# THIS GRAPH SHOWS THAT TOTAL NUMBER OF ORDERS PLACED BY EACH AGE GROUP

df_new1= df.groupby(["Age Group"], as_index=False)["Orders"].sum().sort_values(by="Orders", ascending=False)

ax = sns.barplot(x="Age Group", y="Orders", data=df_new1)

for bars in ax.containers :
    ax.bar_label(bars)
```

```python
# TOTAL NUMBER OF ORDERS FROM TOP 10 STATE

df_new2= df.groupby(["State"], as_index=False)["Orders"].sum().sort_values(by="Orders", ascending=False)
df_new2.head(10)
sns.set(rc={'figure.figsize':(15,5)})
ax= sns.barplot(x="State", y="Orders",data=df_new2.head(10))

for bars in ax.containers :
    ax.bar_label(bars)
```

```python
## TOTAL AMOUNT/SALES FROM TOP 10 STATES

df_tp_sales= df.groupby(["State"], as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False).head(10)
sns.set(rc={"figure.figsize":(15,5)})
ax=sns.barplot(x="State",y="Amount",data=df_tp_sales)

for bars in ax.containers :
    ax.bar_label(bars)
```

<b> From above graphs we can see that most of orders and total sales are from Up, Maharashtra, Karnataka</b>

```python
# numbr of married and unmarried
MS=sns.countplot(x="Marital_Status", hue="Gender", data=df)
sns.set(rc={"figure.figsize":(6,5)})
for bars in MS.containers :
    MS.bar_label(bars)
```

```python
# TOTAL AMOUNT OF SALES BASED ON MARITAL STATUS AS WELL AS GENDER
df_ms_sales=df.groupby(["Marital_Status","Gender"], as_index=False)["Amount"].sum().sort_values(by="Amount",
ascending=False)
sns.set(rc={"figure.figsize":(8,5)})
plot=sns.barplot(x="Marital_Status",y="Amount",hue="Gender", data=df_ms_sales)
for bars in plot.containers :
    plot.bar_label(bars)
```

```python
# NUMBER OF PEOPLE INVOLVED IN VARIOUS OCCUPATION
op=sns.countplot(x="Occupation", data=df)
sns.set(rc={"figure.figsize":(20,5)})
for bars in op.containers :
    op.bar_label(bars)
```

```python
# TOTAL SPENDING BASED ON VARIOUS OCCUPATION
df_ts_oc=df.groupby(["Occupation"], as_index=False)["Amount"].sum().sort_values(by="Amount", ascending=False)
os=sns.barplot(x="Occupation",y="Amount", data=df_ts_oc)
sns.set(rc={"figure.figsize":(20,5)})
for bars in os.containers :
    os.bar_label(bars)
```

<b> From the abov graphs we can see the most buyers are from IT Sector, Aviation, Healthcare sector</b>

```python
# NUMBER OF PRODUCT PER CATEGORY
sns.set(rc={"figure.figsize":(27,3)})
pn=sns.countplot(x="Product_Category", data=df)
for bars in pn.containers :
    pn.bar_label(bars)
```

```python
# TOTAL NUMBER OF ORDERS PER PRODUCT CATEGORY
df_to_pc=df.groupby(["Product_Category"], as_index=False)["Orders"].sum().sort_values(by="Orders", ascending=False)
opp=sns.barplot(x="Product_Category",y="Orders", data=df_to_pc)
for bars in opp.containers :
    opp.bar_label(bars)
```

```python
#total number of sales per product category
sns.set(rc={"figure.figsize":(20,3)})
df_to_pc=df.groupby(["Product_Category"], as_index=False)["Amount"].sum().sort_values(by="Amount", ascending=False).head(10)
spp=sns.barplot(x="Product_Category",y="Amount", data=df_to_pc)
for bars in spp.containers :
    spp.bar_label(bars)
```
